***DETAILS OF THE DATASET FEATURES***

**BeneID**: Unique identifier for the beneficiary associated with the claim.

**ClaimID**: Unique identifier for each claim made by a beneficiary.

**ClaimStartDt**:Start date of the claim, i.e., when the medical services were initiated.

**ClaimEndDt**: End date of the claim, i.e., when the medical services were completed.

**Provider**: Unique identifier for the healthcare provider who rendered the medical services.

**InscClaimAmtReimbursed**: The amount of money reimbursed for the insurance claim made by the beneficiary.

**AttendingPhysician**: Unique identifier for the attending physician who provided primary care to the beneficiary.

**OperatingPhysician**:Unique identifier for the operating physician who performed any surgeries or procedures on the beneficiary, if applicable.

**OtherPhysician**: Unique identifier for any other physician involved in the beneficiary's medical treatment, if applicable.

**AdmissionDt**: The date when the beneficiary was admitted to the healthcare facility for medical treatment.

**ClmAdmitDiagnosisCode**:The diagnosis code associated with the beneficiary's admission to the healthcare facility.

**DeductibleAmtPaid**: The amount of money paid by the beneficiary as a deductible for the claim.

**DischargeDt**: The date when the beneficiary was discharged from the healthcare facility.

**DiagnosisGroupCode**: A group code assigned to a set of related diagnosis codes.

**ClmDiagnosisCode_1 to ClmDiagnosisCode_10**: These columns represent the diagnosis codes associated with the beneficiary's claim, with ClmDiagnosisCode_1 being the primary diagnosis code and ClmDiagnosisCode_2 to ClmDiagnosisCode_10 being additional diagnosis codes, if applicable.

**ClmProcedureCode_1 to ClmProcedureCode_6**: These columns represent the procedure codes associated with the beneficiary's claim, with ClmProcedureCode_1 being the primary procedure code and ClmProcedureCode_2 to ClmProcedureCode_6 being additional procedure codes, if applicable.

```python
#check if same phyician is involved in a claim or not
def physician_same(row):
    atten_oper=row["AttendingPhysician"]==row["OperatingPhysician"]
    oper_other=row["OperatingPhysician"]==row["OtherPhysician"]
    atten_other=row["AttendingPhysician"]==row["OtherPhysician"]
    if atten_oper==True and oper_other==True: # atten = oper = other
        return 0
    elif atten_oper==True and oper_other==False: # atten = oper != other
        return 1
    elif atten_oper==False and oper_other==True: # atten != oper = other
        return 2
    else:# atten != oper != other
        return 3
```

```python
#use the function to undestand whether same  physcian was involved or not 
#0=same physcians in  all three scenario(attend,operating,other)
#1=same physcians for atteding and operating but not other
#2=same operating and other but different attending
#3=all different physcians
result=train_inpatientdata[['Provider']]
same_physician=train_inpatientdata.apply(physician_same,axis=1)
result['same_physician']=same_physician
result
```

```python
#Plot to see Distribution of Physician Involvement in Claims
counts = result['same_physician'].value_counts()
labels =  ['All Physicians Different', 'Operating = Other', 'Attending = Operating', 'All Physicians Same']
plt.figure(figsize=(10, 6))
plt.bar(labels, counts)
plt.xlabel('Physician Involvement')
plt.ylabel('Count')
plt.title('Distribution of Physician Involvement in Claims')
plt.savefig('distribution.png')
plt.show()
```

The above bargraph shows that in the in-patient dataset the involvement of various physician (attending,operating,other) is high and not all the claim is performed by same physicians.

```python
#See days difference between two dates
from datetime import datetime
def calculate_days_between(df, start_col, end_col):
    start_dates = pd.to_datetime(df[start_col])
    end_dates = pd.to_datetime(df[end_col])
    delta = end_dates - start_dates
    return delta.dt.days

```

```python
def plot_duration_distribution(df, duration_col,title):
    counts = df[duration_col].value_counts()
    unique_durations_sorted = counts.sort_index()
    plt.bar(unique_durations_sorted.index, unique_durations_sorted.values)
    plt.xlabel('Duration (in days)')
    plt.ylabel('Count')
    plt.title("Distribution of the difference between  " + title +" Durations")
    plt.savefig(title+".png")
    plt.show()
```

FOR CLAIM START AND CLAIM END DATE.
Explore the usual duration taken, and see what most claim takes time

```python
#for claim dates differenece
result['days_between_claims'] = calculate_days_between(train_inpatientdata,'ClaimStartDt','ClaimEndDt')
```

```python
# Assuming you have a DataFrame named 'result' and a column named 'days_between_claims'
plot_duration_distribution(result, 'days_between_claims','Claim Start and End')
```

FOR ADMISSSION START AND DISCHARGE DATE.
Explore the usual duration taken, and see what most claim takes time.

```python
#for days difference in admission and discharge
result['days_between_admission'] = calculate_days_between(train_inpatientdata,'AdmissionDt','DischargeDt')
```

```python
# Assuming you have a DataFrame named 'result' and a column named 'days_between_admission'
plot_duration_distribution(result, 'days_between_admission','Admission and Discharge')
```

```python
##Check whether the claims are made within the days when you hve taken the service
count1 = result['days_between_claims'].value_counts()
unique_claimdurations_sorted = count1.sort_index()
count2 = result['days_between_admission'].value_counts()
unique_admissiondurations_sorted = count2.sort_index()
plt.figure(figsize=(20,10))
plt.xlabel('Duration (in days)')
plt.ylabel('Count')
plt.plot(unique_claimdurations_sorted.index, unique_claimdurations_sorted.values, label = "Claim_duration_difference",linewidth=5)
plt.plot(unique_admissiondurations_sorted.index, unique_admissiondurations_sorted.values, label = "Admission_duration_difference",linewidth=4)
plt.legend()
plt.savefig('Plotdifference.png')
```

There is difference in the claimdays difference count and admissiondays difference count which means  there are certain claims  which are not within the admission time.
