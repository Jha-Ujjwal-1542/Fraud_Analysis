```python
train_outpatient_df['ClaimStartDt'] = pd.to_datetime(train_outpatient_df['ClaimStartDt'], format="%Y-%m-%d")
train_outpatient_df['ClaimStartDt']
```

```python
train_outpatient_df['ClaimEndDt']
```

```python
train_outpatient_df['ClaimEndDt'] = pd.to_datetime(train_outpatient_df['ClaimEndDt'], format="%Y-%m-%d")
train_outpatient_df['ClaimEndDt']
```

```python
train_outpatient_df['Claim_Duration'] = train_outpatient_df['ClaimEndDt'] - train_outpatient_df['ClaimStartDt']
train_outpatient_df['Claim_Duration'] = train_outpatient_df['Claim_Duration'].dt.days
train_outpatient_df['Claim_Duration']
```

```python
train_outpatient_df['Claim_Duration'].unique()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(train_outpatient_df['Claim_Duration'], bins='auto', color='blue', alpha=0.7, rwidth=0.85)

# Customize the plot
plt.grid(axis='y', alpha=0.5)
plt.xlabel('Claim duration in days')
plt.ylabel('Frequency')
plt.title('Claim duration distribution')

# Display the plot
plt.show()
```

```python
# Using a context manager for plotting
with plt.style.context('seaborn'):
    plt.hist(train_outpatient_df['Claim_Duration'], bins='auto', color='blue', alpha=0.7, rwidth=0.85)
    
    # Customize the plot
    plt.grid(axis='y', alpha=0.5)
    plt.xlabel('Claim duration in days')
    plt.ylabel('Frequency')
    plt.title('Claim duration distribution')

    # Display the plot
    plt.show()
```
